```python
# first import the libraries
import pandas as pd
import numpy as np
# we will use "geopy" library for the coordinates to city name
```

```python
# installing geopy library

# !pip install geopy
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from geopy.extra.rate_limiter import RateLimiter
```

```python
# import data

df = pd.read_csv(r"E:\Projects\SQL projects\50 miles race data analysis project\race_50mile.csv")
```

```python
df.shape
```

```python
# combine first annd last name

df['Full_Name'] = df.First+ ' ' +df.Last
df.head()
```

```python
# create total minutes
# we will have to convert the Time column to the time format

df['Time'] = df['Time'].str.replace('.', ':')

df['Time'] = pd.to_timedelta(df['Time'])
```

```python
df['Total_Minutes'] = (df.Time.dt.total_seconds()/60).round(2)

df.head()
```

```python
df.dtypes
```

```python
# rename division to gender

df.rename(columns = {'Division': 'Gender'}, inplace = True)
df.head()
```

```python
# Need to put city name in the lan and lat columns
def get_lat_long(city, state):
    address = f'{city}, {state}'
    try:
        geolocator = Nominatim(user_agent = "running", timeout = 5)
        location = geolocator.gehttp://localhost:8888/notebooks/Desktop/Python%20notebooks/race_50miles.ipynb#ocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        return None, None
```

```python
# df['Latitude'], df['Longitude'] = zip(df.apply(lambda x: get_lat_long(x['City'], x['State']), axis = 1))
```

```python
df[['Latitude', 'Longitude']] = df.apply(lambda x: pd.Series(get_lat_long(x['City'], x['State'])), axis=1)
```

```python
df['LatLong'] = df.Latitude.astype(str) + ','+ df.Longitude.astype(str)
df.head()
```
